```python
import findspark
findspark.init()
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("example").getOrCreate()
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
names_df=spark.read.option('header',True).option('inferschema',True).csv('/home/labuser/Downloads/Baby_Names.csv')
```

```python
names_df.show()
```

```python
names_df.groupBy('Year').count().sort(col('Year').).show()
```

```python
names_df.write.mode('overwrite').partitionBy('Year').option('path','/home/labuser/raw/processed_data/baby/partitionbyyear').saveAsTable('babynames_year')
```

```python
spark.sql('select * from babynames_year where year=2007').show()
```

```python
names_df.write.mode('overwrite').partitionBy('Year','Sex').option('path','/home/labuser/raw/processed_data/baby/partitionbyyear&gender').saveAsTable('babynames_year_gender')
```

```python
names_df.write\
.mode("overwrite")\
.option("maxRecordsPerFile",4000)\
.partitionBy("Year")\
.option("path","/home/labuser/1ShellPython/raw/processed_data/baby/partitionbyyear_maxi4k")\
.saveAsTable("babynames_year_maxi4k")
```
